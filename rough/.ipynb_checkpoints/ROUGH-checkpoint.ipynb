```python
def zero_pad(inp,out_shape):
    
    m,n,c = inp.shape
    out_m, out_n, out_c = out_shape
    
    to_pad_m = max(out_m - m, 0)
    to_pad_n = max(out_n - n,0)
    to_pad_c = max(out_c - c,0)

    print (to_pad_m, to_pad_n, to_pad_c)
    pad_m1 = to_pad_m//2
    pad_m2 = to_pad_m - pad_m1

    pad_n1 = to_pad_n//2
    pad_n2 = to_pad_n - pad_n1
    
    pad_c1 = to_pad_c//2
    pad_c2 = to_pad_c - pad_c1

    print (pad_m1, pad_m2, pad_n1, pad_n2)
    inp = np.pad(inp, ((pad_m1,pad_m2),(pad_n1,pad_n2),(pad_c1,pad_c2)), 'constant')
    return inp #inp.shape

# out_shape = [6,6,3]
# out = pad(inp=A,out_shape=out_shape)
# print (out)
```

```python


import numpy as np
import tensorflow as tf



def do(init_lr, decay_rate, batch_size = 128, decay_steps = 1000):
    tf.reset_default_graph()
    globalStep = tf.Variable(0, dtype=tf.float32)
    l_rate = tf.train.exponential_decay(init_lr,
                                      globalStep * batch_size,  # Used for decay computation
                                      decay_steps,  # Decay steps
                                      decay_rate,  # Decay rate
                                      staircase=True)  # Will decay the learning rate in discrete interval
    x = tf.placeholder(dtype=tf.float32, shape=[batch_size,100])
    y = tf.placeholder(dtype=tf.float32, shape=[batch_size,2])
    w = tf.get_variable(shape=(100,2), initializer=tf.random_normal_initializer(stddev=0.01), name='w')
    X = tf.matmul(x, w)
    lossCE = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=X, labels=y))
    optimizer = (tf.train.AdamOptimizer(learning_rate=l_rate)
                             .minimize(lossCE, global_step=globalStep))
    l_rate_arr = []
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())

        for i in range(0,3):
            dataX = np.random.random((10000,100))
            dataY = np.random.random((10000,2))

            for j in range(0,10000//batch_size):
                feed_dict = {x:dataX[j*batch_size:(j*batch_size)+batch_size], y:dataY[j*batch_size:(j*batch_size)+batch_size]}
                lr_rate, _ = sess.run([l_rate, optimizer], feed_dict = feed_dict)
                l_rate_arr.append(lr_rate)
                
    return l_rate_arr

```

```python
l_rate_arr1 = do(init_lr=0.0005, decay_rate=0.90, batch_size = 128, decay_steps = 1000)
l_rate_arr2 = do(init_lr=0.0005, decay_rate=0.90, batch_size = 128, decay_steps = 2000)
l_rate_arr1 = do(init_lr=0.0005, decay_rate=0.95, batch_size = 128, decay_steps = 1000)
l_rate_arr2 = do(init_lr=0.0005, decay_rate=0.95, batch_size = 128, decay_steps = 2000)
```

```python
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline

fig1, ax = plt.subplots(1,4,figsize=(40,10), facecolor='y', edgecolor='k')
axs = ax.ravel()
print (axs)
axs[0].plot(l_rate_arr1)
axs[1].plot(l_rate_arr2)
axs[2].plot(l_rate_arr2)
axs[3].plot(l_rate_arr2)
```
